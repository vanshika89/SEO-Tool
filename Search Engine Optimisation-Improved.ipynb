# Search Engine Optimisation Tool (Using Web Scraping)
Web Scraping is an automatic way to retrieve unstructured data from a website and store them in a structured format. Web scraping is used extensively in the area of SEO.

SEO monitoring: Search Engine Optimization is the optimization of the visibility and ranking of a website among different search engines like Google, Yahoo, Bing, etc. Web scraping is used to understand the ranking of the content over time.

```python
# importing all the required libraries for this project
# Requests: It access website's HTML contents or API by sending Get or Post requests.
# Beautiful Soup: It parses HTML or XML documents into a readable format.
import requests
import bs4 
import nltk
import wordcloud
import matplotlib.pyplot as plt
import re
```

### Step 1: Downloading Contents from Web Pages
In this step, a web scraper will download the requested contents from the web page.

```python
try:
    res = requests.get(input('Enter the web url:'))
except Exception as e:
    print('Please enter valid url')
else:
    v= res.text
```

### Step 2: Extracting Data
The data on websites is HTML and mostly unstructured. Hence, in this step, web scraper will parse and extract structured data from the downloaded contents.

```python
soup = bs4.BeautifulSoup(v,'html.parser')
for script in soup(['script','style']):
        script.extract()
        text = soup.get_text()
```

### Step 3: Cleaning the data
Clean the data to easily extract meaningful information. 

```python
from nltk.tokenize import word_tokenize
w_t = word_tokenize(text)
from nltk.corpus import stopwords
s_w = stopwords.words('English')
l = [i for i in w_t if i not in s_w]
c = ' '
text1 = c.join(l)
#The re.sub() method performs global search and global replace on the given string.
res1 = re.sub(r'[^\w\s\d]', '', text1) 
l1 = word_tokenize(res1)
```

### Step 4: Analyzing the Data
After all these steps are successfully done, the web scraper will analyze the data thus obtained.
##### Plotting the frequency distribution Graph

```python
from nltk.probability import FreqDist
fd=FreqDist(l1)
fd.plot(10)
```

##### Making a word cloud for a pictorial representation 

```python
from wordcloud import WordCloud,STOPWORDS
wc=WordCloud().generate(res1)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
